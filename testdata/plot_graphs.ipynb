```python
# import packages
import numpy as np
import changefinder
import matplotlib.pyplot as plt
```

```python
# read timeseries data
data = np.loadtxt("./data.out")
```

```python
# culc scores from changefinder in python
cf = changefinder.ChangeFinder(r=0.01, order=1, smooth=7)

ret_py = []
for i in data:
    score = cf.update(i)
    ret_py.append(score)
```

```python
# plot python scores
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(ret_py)
ax2 = ax.twinx()
ax2.plot(data,'r')
plt.show()
```

```python
# read golang returns
ret_go = np.loadtxt("./ret.out")
```

```python
# plot golang scores
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(ret_go)
ax2 = ax.twinx()
ax2.plot(data,'r')
plt.show()
```
